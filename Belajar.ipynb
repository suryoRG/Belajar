```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data with some missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age": [30, 24, None, 35, 29],
    "Salary": [70000, None, 52000, 100000, 75000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display DataFrame to check initial content
print("DataFrame:")
print(df)

# Handle missing values by dropping rows with missing 'Age' or 'Salary'
df_cleaned = df.dropna(subset=['Age', 'Salary']).copy()

# Display cleaned DataFrame
print("\nCleaned DataFrame:")
print(df_cleaned)

# Ensure 'Age' and 'Salary' are of numerical type using .loc
df_cleaned.loc[:, 'Age'] = df_cleaned['Age'].astype(int)
df_cleaned.loc[:, 'Salary'] = df_cleaned['Salary'].astype(float)

# Display DataFrame types to ensure correct types
print("\nDataFrame dtypes:")
print(df_cleaned.dtypes)

# Check for missing values in 'Age' and 'Salary'
print("Checking for missing values in 'Age' and 'Salary':")
print(df_cleaned[['Age', 'Salary']].isnull().sum())

# Group by 'Age' and calculate the mean 'Salary'
print("\nAttempting to group by 'Age' and calculate mean 'Salary':")

try:
    # Perform groupby operation and inspect the grouped object
    grouped = df_cleaned.groupby('Age')
    print("\nGrouped object:")
    print(grouped)

    # Apply mean manually to check intermediate results
    mean_salary_by_age = grouped['Salary'].mean()
    print("\nMean Salary by Age:")
    print(mean_salary_by_age)
    
    # Plotting
    mean_salary_by_age.plot(kind='bar')
    plt.xlabel('Age')
    plt.ylabel('Average Salary')
    plt.title('Average Salary by Age Group')
    plt.show()
except Exception as e:
    print(f"Groupby Error: {e}")
```
